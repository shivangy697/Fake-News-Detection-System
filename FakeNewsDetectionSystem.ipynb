```python
data_merge = pd.concat([data_fake, data_true], axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject','date'], axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac =1)
data.head()
```

```python
data.reset_index(inplace =True)
data.drop(['index'], axis =1, inplace =True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
data['text'] =data['text'].apply(wordopt)
```

```python
x = data['text']
y =data['class']
```

```python
x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size =0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate the TfidfVectorizer
vectorization = TfidfVectorizer()

# Fit and transform the training data
xv_train = vectorization.fit_transform(x_train)

# Transform the test data
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Instantiate the GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state=0)

# Fit the model using the training data
GB.fit(xv_train, y_train)
```

```python
pred_gb =GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_label(n):
    return "Not Fake News!" if n == 1 else "Fake News!"

def manual_testing(news):
    processed_news = wordopt(news)  # Process the input news string
    transformed_news = vectorization.transform([processed_news])  # Transform the processed news once

    # Get predictions from all models
    predictions = {
        "LR Prediction": LR.predict(transformed_news)[0],
        "DT Prediction": DT.predict(transformed_news)[0],
        "GBC Prediction": GB.predict(transformed_news)[0],
        "RFC Prediction": RF.predict(transformed_news)[0]
    }

    # Generate and print output labels for each model
    for model, prediction in predictions.items():
        print(f"{model}: {output_label(prediction)}")

# Usage
manual_testing("Your news text here")
```
